# **Data Visualization**

## **Visualization No.1: Top 10 Countries by Number of Medals**


```python
# Get the top 10 countries by total medals using head()
top_10_countries = df.head(10)

# Create a bar chart for the top 10 countries
plt.figure(figsize=(12, 4))
plt.bar(top_10_countries['Country'], top_10_countries['Total Medal'], color='gold')
plt.title('Top 10 Countries by Number of Medals')
plt.xlabel('Country')
plt.ylabel('Total Medals')
plt.xticks(rotation=90)
plt.grid(axis='y')  # Add gridlines for better readability
plt.show()
```

**Insight**
The bar chart highlights the top 10 countries by total medals, showcasing their Olympic success 

## **Visualization No.2: Bottom 10 Countries by Number of Medals**


```python
# Get the top 10 countries by total medals using head()
bottom_10_countries = df.tail(10)

# Create a bar chart for the top 10 countries
plt.figure(figsize=(12, 4))
plt.bar(bottom_10_countries['Country'], bottom_10_countries['Total Medal'], color='black')
plt.title('Bottom 10 Countries by Number of Medals')
plt.xlabel('Country')
plt.ylabel('Total Medals')
plt.xticks(rotation=90)
plt.grid(axis='y')  # Add gridlines for better readability
plt.show()
```

**Insight**
This bar chart highlights the bottom 10 countries by total medals, revealing those with the least Olympic success
and indicating potential areas for improvement in sports development.

## **Visualization No.3:Pie Chart for Medal Distribution**

```python
# Calculate total medals for each type (Gold, Silver, Bronze) across all countries
total_gold = df['Gold'].sum()
total_silver = df['Silver'].sum()
total_bronze = df['Bronze'].sum()

# Prepare data for the pie chart
medal_distribution = [total_gold, total_silver, total_bronze]
medal_labels = ['Gold', 'Silver', 'Bronze']
colors = ['gold', 'silver', '#cd7f32']  # Colors for each medal type

# Create the pie chart
plt.figure(figsize=(4, 4))
plt.pie(medal_distribution, labels=medal_labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title('Medal Distribution Among All Countries')
plt.axis('equal')  # Equal aspect ratio ensures the pie chart is circular
plt.show()
```

**Insight**
This pie chart illustrates the overall medal distribution among gold, silver, and bronze categories, 
showcasing the proportion of each type in the total medal count.

## **Visualization No.4:Pie Chart for Top 10 Countries by Total Medals**

```python
# Sort the DataFrame by 'Total Medal' in descending order and use head() to get top 10 countries
top_10_countries = df.sort_values(by='Total Medal', ascending=False).head(10)

# Define custom named colors for each country
colors = ['DarkCyan', 'Crimson', 'DodgerBlue', 'ForestGreen', 'HotPink', 'GoldenRod', 
          'MediumPurple', 'DarkOrange', 'SkyBlue', 'MediumSeaGreen']

# Create a pie chart for the top 10 countries
plt.figure(figsize=(6, 7))
plt.pie(top_10_countries['Total Medal'], labels=top_10_countries['Country'], colors=colors, autopct='%1.1f%%', startangle=140)
plt.title('Top 10 Countries by Total Medal Distribution')
plt.axis('equal')  # Equal aspect ratio ensures that the pie is drawn as a circle
plt.show()
```

**Insight**
This pie chart represents the distribution of total medals among the top 10 countries, allowing for a quick comparison of their relative success in the Olympics.

## **Visualization No.5:Medal Breakdown: Top 10 Countries by Gold, Silver, and Bronze Medals**

```python
# Sort the data to get the top 10 countries with the highest number of Gold, Silver, and Bronze medals in descending order
# by using sort_values for each medal type and retrieving the top 10 with head(10) for further analysis.

top_10_gold = df.sort_values(by='Gold', ascending=False).head(10)
top_10_silver = df.sort_values(by='Silver', ascending=False).head(10)
top_10_bronze = df.sort_values(by='Bronze', ascending=False).head(10)

# Create a figure and subplots with 3 charts (1 row, 3 columns)
fig, axes = plt.subplots(1, 3, figsize=(18, 6)) # Set the figure size to 18 inches wide and 6 inches tall to ensure the subplots are large and clearly visible.

# Subplot 1: horizontal bar chart for Gold medals
axes[0].barh(top_10_gold['Country'], top_10_gold['Gold'], color='gold')
axes[0].set_title('Top 10 Countries by Gold Medals')
axes[0].set_xlabel('Number of Gold Medals')
axes[0].invert_yaxis()  # To have the country with most medals at the top

# Subplot 2: horizontal bar chart for Silver medals
axes[1].barh(top_10_silver['Country'], top_10_silver['Silver'], color='silver')
axes[1].set_title('Top 10 Countries by Silver Medals')
axes[1].set_xlabel('Number of Silver Medals')
axes[1].invert_yaxis()  # To have the country with most medals at the top

# Subplot 3: horizontal bar chart for Bronze medals
axes[2].barh(top_10_bronze['Country'], top_10_bronze['Bronze'], color='#cd7f32')  # Bronze color
axes[2].set_title('Top 10 Countries by Bronze Medals')
axes[2].set_xlabel('Number of Bronze Medals')
axes[2].invert_yaxis()  # To have the country with most medals at the top

plt.tight_layout() # Adjust layout for better spacing
plt.show() # Show the plots
```

**Insight**
This visualization effectively compares the top 10 countries in terms of Gold, Silver, and Bronze medals won at the Olympics. 
By displaying each medal type side-by-side, it highlights the strengths and performance differences among these leading nations, enabling viewers to easily identify which countries excelled in specific medal categories.

## **Visualization No.6:Subplots for Countries Without Gold Medals: Comparison of Silver and Bronze**

```python
# Filter the DataFrame for countries with zero gold medals
no_gold_countries = df[df['Gold'] == 0]

# Create subplots for silver and bronze medals
fig, axs = plt.subplots(1, 2, figsize=(14, 8))

# Silver Medals Bar Chart
axs[0].bar(no_gold_countries['Country'], no_gold_countries['Silver'], color='darkgray')
axs[0].set_title('Silver Medals by Countries Without Gold Medals')
axs[0].set_xlabel('Country')
axs[0].set_ylabel('Silver Medals')
axs[0].set_xticklabels(no_gold_countries['Country'], rotation=90)  # Rotate x-tick labels for better readability
axs[0].grid(axis='y', linestyle='--', alpha=0.7)  # Add gridlines

# Bronze Medals Bar Chart
axs[1].bar(no_gold_countries['Country'], no_gold_countries['Bronze'], color='#CD7F32')  # Bronze color
axs[1].set_title('Bronze Medals by Countries Without Gold Medals')
axs[1].set_xlabel('Country')
axs[1].set_ylabel('Bronze Medals')
axs[1].set_xticklabels(no_gold_countries['Country'], rotation=90)  # Rotate x-tick labels for better readability
axs[1].grid(axis='y', linestyle='--', alpha=0.7)  # Add gridlines
plt.tight_layout() # Automatically adjust subplot parameters for better layout
plt.show()
```

**Insight**
This visualization effectively compares the number of silver and bronze medals among countries that have not secured any gold medals, illustrating their overall performance and competitiveness in Olympic events.

## **Visualization No.7:Comparison of Total Male and Female Athletes**

```python
# Calculate the total number of male and female athletes
total_male = df['Male Athletes'].sum()  # Sum of all male athletes
total_female = df['Female  Athletes'].sum()  # Sum of all female athletes

# Prepare data for plotting
genders = ['Male', 'Female']  # Gender labels
counts = [total_male, total_female]  # Corresponding counts

# Create a bar chart to compare male and female athletes
plt.figure(figsize=(4, 4))  # Set the figure size for better visibility
bars = plt.bar(genders, counts, color=['blue', 'pink'])  # Create the bar chart

# Adding titles and labels
plt.title('Total Count of Male vs Female Athletes')  # Title of the chart
plt.xlabel('Gender')  # X-axis label
plt.ylabel('Number of Athletes')  # Y-axis label

# Display the count on top of the bars using the bar heights directly
plt.bar_label(bars , fontsize=8)  # Use plt.bar_label to display the counts on top of the bars
plt.tight_layout()  # Adjust layout to make it look better
plt.show()  # Display the chart
```

**Insight**
The total number of male athletes (5,077) is slightly higher than female athletes (5,041), indicating a nearly balanced gender representation in the Olympics. This reflects progress towards gender equality, but there is still room for improvement in promoting female participation in sports.

## **Visualization No.8:Medal Winning Rate for All Countries**

```python

# Calculate Medal Winning Rate
df['Winning Rate'] = df['Total Medal'] / df['Total  Athletes']

# Bar chart for Medal Winning Rate
plt.figure(figsize=(18, 6))
plt.bar(df['Country'], df['Winning Rate'], color='skyblue')
plt.xlabel('Country')
plt.ylabel('Winning Rate (Total Medals per Athlete)')
plt.title('Medal Winning Rate by Country')
plt.xticks(rotation=90 ,fontsize=9 )
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

**Insight**
This bar chart shows the Medal Winning Rate for different countries. It helps compare how well each country turns athlete participation into medals. A higher winning rate means a country is more successful in winning medals relative to the number of athletes it has.

## **Visualization No.9:Heatmap of Medal Types for Top 10 Ranked Countries**

```python
# Sort the data by rank and select the top 10 countries
top_10_countries = df.sort_values(by='Rank').head(10)

# Prepare the data for gold, silver, and bronze medal counts
medal_types_data_top_10 = top_10_countries[['Country', 'Gold', 'Silver', 'Bronze']].set_index('Country')

# Create a heatmap for the top 10 ranked countries' medal types
plt.figure(figsize=(12, 6))  # Set the size of the figure
plt.imshow(medal_types_data_top_10, cmap='viridis', aspect='auto')  # Use 'viridis' colormap
plt.colorbar(label='Count of Medals')  # Add a color bar to indicate medal counts
plt.xticks(ticks=np.arange(len(medal_types_data_top_10.columns)), labels=medal_types_data_top_10.columns)  # Set x-tick labels
plt.yticks(ticks=np.arange(len(medal_types_data_top_10.index)), labels=medal_types_data_top_10.index)  # Set y-tick labels
plt.title('Heatmap of Medal Types for Top 10 Ranked Countries')  # Set the title of the heatmap
plt.show()  # Display the heatmap
```

**Insight**
The heatmap visually represents the distribution of gold, silver, and bronze medals among the top 10 ranked countries, highlighting which countries excelled in specific medal types. It allows for quick
comparisons, revealing performance trends and potential areas for improvement in athletic strategies. This analysis can guide future training and resource allocation to enhance Olympic outcomes.
